ACf and Pacf plots:
these plots are most used plots in time series these plots summarize graphically the strength between varibale and time with observarions at pirior time steps.
==>the diffrence between acf and pacf plots are difficult and confusing for begginers to timeseries forecasting
1. Here we calculate the correlaion for Time series with the observarions of previous time steps called Lags
2.The correlation of the timeseries  observations are calculated with values  of the same series with the previous times, this is called serial correlation or auto correlation
3.ACF which contains with lags like==> acf(lags=some value) called ==> Auto Correlation Function or some times called  as correlogram
4. here we use plot_acf() form the statsmodels library



PACF ==> Partial AutoCorrelation Function
A partial auto Correlation is a summary of relationship between  an observation in a  time  series with observations at pirior time steps with the relationships of intervening observations removed

```python
#ACF
from statsmodels.graphics.tsaplots import plot_acf
df= pd.read_csv(r"C:\Users\RaviKiran\OneDrive\Desktop\Data Science\Data Sets\daily-minimum-temperatures.csv", header=0, index_col=0)
plot_acf(df,lags= 40)# -1 to +1
pyplot.show()
# the points that crosses the shaded reagion which is std which means it is significant varibale
```
